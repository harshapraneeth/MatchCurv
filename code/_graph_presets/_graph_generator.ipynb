# Graph Generator

This script for creating random graphs using the networkx library.

```python
from matplotlib import pyplot as plt
from tqdm import tqdm
import networkx as nx
import numpy as np
import random
import pickle
```

```python
labels = [
    "127.0.0.%d" % i
    for i in range(1, 255)
]
```

```python
preset = "temp.b"
graph: nx.Graph = nx.connected_watts_strogatz_graph(100, 8, 0.33)
graph = nx.relabel_nodes(graph, {i: label for i, label in enumerate(labels)})

pos = nx.spring_layout(graph)
plt.figure(figsize = (10, 10))
nx.draw(graph, pos = pos, with_labels = False)

with open(preset, "wb") as file:
    pickle.dump((graph, pos), file)

print(graph.number_of_nodes(), graph.number_of_edges(), nx.is_connected(graph))
```
